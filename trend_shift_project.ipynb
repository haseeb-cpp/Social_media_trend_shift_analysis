```python
#Social media text is often messy:

import re

def clean_text(text):
    text = str(text).lower()                        # lowercase
    text = re.sub(r"http\S+|www.\S+", "", text)     # remove links
    text = re.sub(r"@\w+", "", text)                # remove mentions
    text = re.sub(r"#", "", text)                   # remove hashtag symbol (keep word)
    text = re.sub(r"[^a-zA-Z\s]", "", text)         # remove special chars, numbers, emojis
    text = re.sub(r"\s+", " ", text).strip()        # remove extra spaces
    return text

book1["clean_text"] = book1["Text"].apply(clean_text)
tweets["clean_text"] = tweets["text"].apply(clean_text)
```

```python
import pandas as pd

# Load dataset
tweets = pd.read_csv("Tweets.csv")

# Preview
print(tweets.head())
print(tweets.columns)

# 1. Convert tweet_created → datetime
tweets["tweet_created"] = pd.to_datetime(tweets["tweet_created"], errors="coerce")

# 2. Extract useful time parts
tweets["Year"] = tweets["tweet_created"].dt.year
tweets["Month"] = tweets["tweet_created"].dt.month
tweets["Day"] = tweets["tweet_created"].dt.day
tweets["Hour"] = tweets["tweet_created"].dt.hour
tweets["DayOfWeek"] = tweets["tweet_created"].dt.day_name()
tweets["MonthName"] = tweets["tweet_created"].dt.strftime("%B")

# 3. Handle text/hashtags (hashtags extracted from text)
tweets["Hashtags"] = tweets["text"].str.findall(r"#\w+")
tweets["Hashtags"] = tweets["Hashtags"].apply(
    lambda x: " ".join(x).lower() if isinstance(x, list) else ""
)

# 4. Handle nulls
tweets = tweets.dropna(subset=["text"])  # drop rows with missing text
tweets = tweets.fillna({"tweet_location": "Unknown", "user_timezone": "Unknown"})

# 5. Standardize numeric fields
tweets["retweet_count"] = pd.to_numeric(tweets["retweet_count"], errors="coerce").fillna(0).astype(int)

print("Tweets dataset cleaned ✅")
print(tweets.head())
```

```python
#Step-1.0 Imports + assumptions

#We’ll normalize timestamps, build daily & monthly timelines, smooth them, detect spikes, and quickly summarize what caused the spikes.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Work on a copy to avoid SettingWithCopy warnings
df = master_df.copy()


#Why: keeps your original master_df intact.
```

```python
# Step1.1) Timestamp normalization (one-time guard)

#Fix mixed formats/timezones.

#Keep only rows with a valid timestamp for time-based charts.

# Normalize timestamps: parse → force UTC → drop timezone (naive)
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce", utc=True)
df["Timestamp"] = df["Timestamp"].dt.tz_convert(None)

# Keep only rows that actually have a time
t = df.dropna(subset=["Timestamp"]).copy()


#Why: makes .resample() and .min()/.max() reliable and avoids tz errors.
```

```python
#Step1.2) Daily timeline (+ 7-day smoothing) and spike detection

#We count posts per day, smooth with a 7-day moving average, then flag unusually high days (z-score > 2.5).

# ---- Daily post volume ----
daily = (
    t.set_index("Timestamp")
     .resample("D")          # daily
     .size()                 # count of rows per day
     .rename("posts")
)

# 7-day moving average (smooths noise)
daily_ma7 = daily.rolling(window=7, min_periods=1).mean()

# Spike detection via z-score
mean, std = daily.mean(), daily.std(ddof=0)
zscore = (daily - mean) / (std if std != 0 else 1)  # avoid divide-by-zero
spikes = zscore[zscore > 2.5].sort_values(ascending=False)  # tune threshold if needed
top_spike_days = spikes.head(5).index  # show top 5 spikes

# Plot: daily + MA(7) with spike markers
plt.figure(figsize=(12, 5))
plt.plot(daily.index, daily.values, label="Daily posts", linewidth=1)
plt.plot(daily_ma7.index, daily_ma7.values, label="7-day moving average", linewidth=2)
for d_ in top_spike_days:
    plt.axvline(d_, alpha=0.15)  # light marker for spikes
plt.title("Daily Post Volume (with 7-day Smoothing)")
plt.xlabel("Date"); plt.ylabel("Posts"); plt.legend(); plt.tight_layout()
plt.show()

print("Top spike days (z-score > 2.5):")
print(spikes.head(5))


#Why:Daily granularity shows short-term volatility.

#7-day MA helps reveal the underlying trend.

#Spike detection gives you dates to investigate “what happened here?”
```

```python
# Plot: daily + MA(7) with clear spike highlights
plt.figure(figsize=(14, 6))

# Raw daily posts (light grey, thin line)
plt.plot(daily.index, daily.values, 
         color="lightgrey", linewidth=1, label="Daily posts")

# Smoothed 7-day moving average (bold blue line)
plt.plot(daily_ma7.index, daily_ma7.values, 
         color="royalblue", linewidth=2.5, label="7-day moving average")

# Mark spikes with red dots
plt.scatter(top_spike_days, daily.loc[top_spike_days], 
            color="red", s=80, zorder=5, label="Detected spikes")

# Annotate spike values
for d_ in top_spike_days:
    plt.text(d_, daily.loc[d_] + 5,   # shift label slightly above point
             f"{daily.loc[d_]}", 
             color="red", fontsize=9, ha="center")

# Formatting
plt.title("Daily Post Volume with Trend and Spikes", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Posts", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#Step 1.3- Monthly timeline (+ long-term growth)

#Roll up to month level.

#Compute simple trend slope (is volume growing or shrinking?).

#(Optional) YoY% for a quick growth feel.

# ---- Monthly post volume ----
monthly = (
    t.set_index("Timestamp")
     .resample("MS")         # Month Start; cleaner axis labels
     .size()
     .rename("posts")
)

# Simple long-term trend (slope of a line fit)
if len(monthly) >= 2:
    x = np.arange(len(monthly))
    slope = np.polyfit(x, monthly.values, 1)[0]  # posts per month change
else:
    slope = np.nan

# Optional: Year-over-Year % change (needs at least 13 months)
monthly_yoy = monthly.pct_change(12) * 100

# Plot monthly posts
plt.figure(figsize=(12, 5))
plt.plot(monthly.index, monthly.values, linewidth=2)
plt.title("Monthly Post Volume")
plt.xlabel("Month"); plt.ylabel("Posts"); plt.tight_layout()
plt.show()

print(f"Monthly trend slope (≈ posts added per month): {slope:.2f}")
if monthly_yoy.notna().any():
    print("\nRecent YoY% changes (last few months):")
    print(monthly_yoy.dropna().tail(6).round(2))


#Why:Monthly view reveals long-term trend shifts.

#Slope > 0 → growing activity; < 0 → declining.

#YoY% highlights seasonality and structural changes.
```

```python
#Step 1.4) Quick spike attribution (what caused the spikes?)

#For each spike date, show top platforms and hashtags that day.
#(Hashtags are split on whitespace; adjust if yours are comma-separated.)

def summarize_spike(day, window_days=1):
    start = pd.Timestamp(day)
    end = start + pd.Timedelta(days=window_days)  # [day, day+window)
    slice_ = t[(t["Timestamp"] >= start) & (t["Timestamp"] < end)].copy()

    # By platform
    by_platform = slice_["Platform"].value_counts().head(5)

    # Top hashtags
    if "Hashtags" in slice_.columns:
        tags = (
            slice_["Hashtags"].dropna().astype(str)
                  .str.split()           # split on whitespace
                  .explode()
                  .str.lower()
        )
        top_tags = tags.value_counts().head(10)
    else:
        top_tags = pd.Series(dtype=int)

    return by_platform, top_tags

# Print a short summary for the top spike days
for d_ in top_spike_days:
    plat, tags = summarize_spike(d_)
    print(f"\n=== Spike on {pd.Timestamp(d_).date()} (posts={int(daily.loc[d_])}) ===")
    print("By platform:")
    print(plat)
    print("\nTop hashtags:")
    print(tags)


#Why:Turns “there was a spike” into a story: which platforms and hashtags drove it.
#------------------------------------------------------------------------------------
#What just covered in Step 1

#Daily activity + 7-day smoothing → short-term patterns.

#Monthly activity + trend slope → long-term shifts.

#Spike detection → outlier days.

#Spike attribution → platforms & hashtags behind spikes.

#This gives a solid backbone for the rest of the trend-shift analysis.
```

```python
#This step answers questions like:

#Which platform (Twitter, Book1, Viral) is the most active?

#How has each platform’s contribution changed over time?

#Is one platform growing while another is declining?

# Step 2.1 — Posts per Platform Over Time

#We’ll analyze the number of posts per platform across time.


import matplotlib.pyplot as plt

# Ensure Timestamp is datetime
t["Timestamp"] = pd.to_datetime(t["Timestamp"], errors="coerce")

# Group by Platform + Month
posts_platform = (
    t.set_index("Timestamp")
     .groupby("Platform")
     .resample("ME")   # monthly counts
     .size()
     .unstack("Platform")   # columns = platforms
     .fillna(0)
)

# Plot: Line chart for each platform
plt.figure(figsize=(12,6))
for platform in posts_platform.columns:
    plt.plot(posts_platform.index, posts_platform[platform], label=platform, linewidth=2)

plt.title("Posts per Platform Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Number of Posts")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Alternative: Stacked Area chart (shows share of activity)
posts_platform.plot.area(figsize=(12,6), alpha=0.7)
plt.title("Platform Share of Posts Over Time", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Number of Posts")
plt.legend(title="Platform")
plt.tight_layout()
plt.show()
```

```python
# Step 2.1 – Platform trends over time (cleaned)

# 1. Clean platform names (remove spaces, lowercase, capitalize)
t["Platform"] = t["Platform"].astype(str).str.strip().str.lower().str.capitalize()

# 2. Resample posts per platform (monthly for smoother view)
platform_trends = (
    t.set_index("Timestamp")
     .groupby("Platform")
     .resample("ME")
     .size()
     .reset_index(name="posts")
)

# 3. Pivot for stacked area (share of posts)
platform_pivot = platform_trends.pivot(index="Timestamp", columns="Platform", values="posts").fillna(0)

# 4. Plot: absolute trends per platform
plt.figure(figsize=(12, 5))
sns.lineplot(data=platform_trends, x="Timestamp", y="posts", hue="Platform", linewidth=2)
plt.title("Posts per Platform Over Time (Monthly)")
plt.xlabel("Date"); plt.ylabel("Number of Posts")
plt.legend(title="Platform")
plt.tight_layout()
plt.show()

# 5. Plot: relative share (stacked area)
plt.figure(figsize=(12, 5))
(platform_pivot.div(platform_pivot.sum(axis=1), axis=0) * 100).plot.area(
    figsize=(12, 5), alpha=0.7, cmap="tab20"
)
plt.title("Platform Share of Posts Over Time (%)")
plt.ylabel("Share of Posts (%)")
plt.xlabel("Date")
plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# 6. Check for spikes by year (helps explain anomalies like 2015)
print("\nPosts per year:")
print(t["Timestamp"].dt.year.value_counts().sort_index())
```

```python
# Step 2.2 – Platform Growth Rates (YoY % change)
#Here, instead of just raw counts, we measure how fast each platform is growing or shrinking over time.
#This helps us detect trend shifts (e.g., Instagram overtaking Twitter, Facebook decline, etc.).

# Resample yearly for clearer long-term trend
platform_yearly = (
    t.set_index("Timestamp")
     .groupby("Platform")
     .resample("YE")
     .size()
     .reset_index(name="posts")
)

# Compute year-over-year growth (%) for each platform
platform_yearly["YoY_growth_%"] = (
    platform_yearly.groupby("Platform")["posts"].pct_change() * 100
)

# Plot yearly trends
plt.figure(figsize=(12, 5))
sns.lineplot(data=platform_yearly, x="Timestamp", y="posts", hue="Platform", marker="o")
plt.title("Platform Post Volume per Year")
plt.xlabel("Year"); plt.ylabel("Posts")
plt.tight_layout()
plt.show()

# Plot growth rates
plt.figure(figsize=(12, 5))
sns.lineplot(data=platform_yearly, x="Timestamp", y="YoY_growth_%", hue="Platform", marker="o")
plt.axhline(0, color="gray", linestyle="--")
plt.title("Platform YoY Growth Rate (%)")
plt.xlabel("Year"); plt.ylabel("Growth %")
plt.tight_layout()
plt.show()

# Print summary (last 3 years of growth per platform)
print("\nRecent growth rates (last 3 years):")
print(platform_yearly.groupby("Platform").tail(3))
```

```python
#Step 2.3 – Platform Market Share (Relative Popularity Over Time).

#So far, we’ve looked at absolute counts and growth rates.
#But trend shifts are best seen in terms of share — how much of the total conversation each platform owns at a given time.

# Step 2.3 – Platform Market Share (Relative Popularity)

# Resample monthly for smoother trends
platform_monthly = (
    t.set_index("Timestamp")
     .groupby("Platform")
     .resample("ME")
     .size()
     .reset_index(name="posts")
)

# Compute total posts per month
monthly_totals = (
    platform_monthly.groupby("Timestamp")["posts"].transform("sum")
)

# Calculate share %
platform_monthly["share_%"] = (platform_monthly["posts"] / monthly_totals) * 100

# Plot stacked area chart (clearer view of share shifts)
plt.figure(figsize=(12, 6))
for platform in platform_monthly["Platform"].unique():
    subset = platform_monthly[platform_monthly["Platform"] == platform]
    plt.fill_between(subset["Timestamp"], subset["share_%"], 
                     label=platform, alpha=0.6)

plt.title("Platform Market Share Over Time")
plt.xlabel("Date"); plt.ylabel("Share of Posts (%)")
plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Print last available month’s market shares
latest_shares = (
    platform_monthly[platform_monthly["Timestamp"] == platform_monthly["Timestamp"].max()]
    [["Platform", "share_%"]]
    .sort_values("share_%", ascending=False)
)
print("\nLatest Market Share of Platforms:")
print(latest_shares)
```

```python
#Top 20 Hashtags Overall
#Goal:To find which hashtags appear most frequently across all posts — these represent the core or recurring trends in the dataset.
# Step 3.1 — Top 20 Hashtags Overall
# Step 3.1 — Top 20 Hashtags Overall (Fixed)

# Combine all hashtags from the dataset safely
if "Hashtags" in t.columns:
    all_tags = (
        t["Hashtags"]
        .dropna()
        .astype(str)
        .str.replace(r"[#,\[\]']", "", regex=True)  # remove extra symbols
        .str.lower()
        .str.split()
        .explode()
        .dropna()
        .astype(str)
    )

    # Remove invalid entries (like numbers or very short tokens)
    all_tags = all_tags[all_tags.str.match(r"^[a-zA-Z0-9_]+$")]

    # Count frequency of each hashtag
    top_tags = all_tags.value_counts().head(20)

    # --- Visualization: Horizontal Bar Chart ---
    plt.figure(figsize=(10, 6))
    sns.barplot(y=top_tags.index, x=top_tags.values, hue=top_tags.index, palette="viridis", legend=False)
    plt.title("Top 20 Hashtags Overall", fontsize=14)
    plt.xlabel("Frequency")
    plt.ylabel("Hashtag")
    plt.tight_layout()
    plt.show()

    # --- Word Cloud (Safe from float/NaN issues) ---
    from wordcloud import WordCloud
    wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="plasma").generate(" ".join(all_tags))
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("Word Cloud of Hashtags", fontsize=14)
    plt.show()

    print("\nTop 20 Hashtags:")
    print(top_tags)

else:
    print("⚠️ 'Hashtags' column not found in the dataset.")
```

```python
# Goal:To visualize how frequently the most-used hashtags appear over time — helping detect trend shifts (e.g., “#AI” surging while “#COVID” fades).
# Step 3.2 — Hashtag Trend Over Time

# 1️⃣ Take only the top 10 hashtags from the previous step
top10_tags = top_tags.index[:10]

# 2️⃣ Extract and clean hashtags again, keeping only top10
hashtags_time = (
    t[["Timestamp", "Hashtags"]]
    .dropna()
    .assign(Hashtags=lambda df: df["Hashtags"].astype(str).str.lower().str.replace(r"[#,\[\]']", "", regex=True))
    .assign(Hashtags=lambda df: df["Hashtags"].str.split())
    .explode("Hashtags")
)
hashtags_time = hashtags_time[hashtags_time["Hashtags"].isin(top10_tags)]

# 3️⃣ Resample monthly and count occurrences
hashtags_time = (
    hashtags_time.set_index("Timestamp")
    .groupby("Hashtags")
    .resample("ME")
    .size()
    .reset_index(name="count")
)

# 4️⃣ Plot multi-line chart (each line = one hashtag)
plt.figure(figsize=(12,6))
sns.lineplot(data=hashtags_time, x="Timestamp", y="count", hue="Hashtags", linewidth=2)
plt.title("Monthly Trend of Top 10 Hashtags")
plt.xlabel("Date"); plt.ylabel("Count")
plt.legend(title="Hashtag", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Goal:Detect which hashtags are gaining or losing popularity over time.
#This helps uncover new trends (emerging topics) and fading trends (declining buzz).

# Step 3.3 — Emerging & Declining Hashtags

# 1️⃣ Prepare monthly hashtag counts
hashtag_trends = (
    hashtags_time
    .groupby(["Hashtags", "Timestamp"])["count"]
    .sum()
    .reset_index()
)

# 2️⃣ Compute growth: compare first 3 months vs last 3 months
trend_summary = (
    hashtag_trends.groupby("Hashtags", group_keys=False)
    .apply(
        lambda df: pd.Series({
            "first_avg": df.head(3)["count"].mean(),
            "last_avg": df.tail(3)["count"].mean()
        }),
        include_groups=False  # ✅ avoid deprecation warning
    )
    .reset_index()
)

# 3️⃣ Calculate growth rate (%)
trend_summary["growth_%"] = (
    ((trend_summary["last_avg"] - trend_summary["first_avg"]) /
     (trend_summary["first_avg"] + 1e-6)) * 100
)

# 4️⃣ Separate emerging vs. declining
emerging = trend_summary.sort_values("growth_%", ascending=False).head(10)
declining = trend_summary.sort_values("growth_%", ascending=True).head(10)

# 5️⃣ Visualize (no warnings)
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(y="Hashtags", x="growth_%", hue="Hashtags", data=emerging,
            palette="Greens", legend=False, ax=ax[0])
ax[0].set_title("Top 10 Emerging Hashtags (Highest Growth)")
ax[0].set_xlabel("Growth %"); ax[0].set_ylabel("Hashtag")

sns.barplot(y="Hashtags", x="growth_%", hue="Hashtags", data=declining,
            palette="Reds", legend=False, ax=ax[1])
ax[1].set_title("Top 10 Declining Hashtags (Largest Drop)")
ax[1].set_xlabel("Growth %"); ax[1].set_ylabel("")

plt.tight_layout()
plt.show()

print("\n🚀 Emerging Hashtags:\n", emerging[["Hashtags","growth_%"]])
print("\n📉 Declining Hashtags:\n", declining[["Hashtags","growth_%"]])
```

```python
#Step 4 — Engagement Dynamics
#🎯 Goal:Understand how people interact with content across platforms and over time.
#We’ll analyze likes, retweets, comments, and see which platforms or time windows drive higher engagement.

#Step 4.1 — Average Engagement per Platform

#This gives a baseline view of which platforms generate more engagement on average.

# Check for engagement columns
engagement_cols = [c for c in t.columns if c.lower() in ["likes", "retweets", "shares", "comments", "views"]]

if engagement_cols:
    engagement_summary = (
        t.groupby("Platform")[engagement_cols]
        .mean(numeric_only=True)
        .round(2)
        .sort_values(by=engagement_cols[0], ascending=False)
    )

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=engagement_summary.reset_index().melt(id_vars="Platform", value_name="Avg Value"),
        x="Avg Value", y="Platform", hue="variable",
        palette="coolwarm", legend=True
    )
    plt.title("Average Engagement Metrics per Platform")
    plt.xlabel("Average Value"); plt.ylabel("Platform")
    plt.legend(title="Engagement Type", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    print("Average Engagement per Platform:\n")
    print(engagement_summary)
else:
    print("⚠️ No engagement-related columns found.")
```

```python
#🎯 Goal:Track engagement levels over time to reveal when users were most active or reactive.
#This helps answer questions like:

#“When did engagement peak?”
#“Are people interacting less or more over time?”

# Step 4.2 — Engagement Over Time

# Check if engagement columns exist
engagement_cols = [c for c in t.columns if c.lower() in ["likes", "retweets", "shares", "comments", "views"]]

if engagement_cols:
    # Prepare monthly engagement data
    engagement_time = (
        t.set_index("Timestamp")
         .resample("ME")[engagement_cols]
         .mean(numeric_only=True)
         .reset_index()
    )

    # Melt for plotting (long format)
    engagement_melted = engagement_time.melt(id_vars="Timestamp", var_name="Engagement Type", value_name="Average Value")

    # Plot
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=engagement_melted, x="Timestamp", y="Average Value", hue="Engagement Type", linewidth=2)
    plt.title("Average Engagement Over Time")
    plt.xlabel("Date")
    plt.ylabel("Average Engagement Value")
    plt.legend(title="Engagement Type", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

    # Print latest snapshot
    print("\nLatest Monthly Engagement Averages:")
    print(engagement_time.tail(3))

else:
    print("⚠️ No engagement-related columns found.")
```
